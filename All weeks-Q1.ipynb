```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
%matplotlib inline

df = pd.read_excel('surveys.xlsx')
df_quest = pd.read_excel('surveyquest.xlsx')
new_df = pd.merge(df, df_quest, on='question', how='inner')
print(len(new_df))


is_catg_low = df_quest['category']=='low'
is_catg_high = df_quest['category']=='high'
#df_quest = df_quest[is_catg]
categoryone = len(df_quest[is_catg_low])
categorytwo = len(df_quest[is_catg_high])
print(categoryone)
print(categorytwo)

plt.bar(x ='category', height = 'question', data = df_quest, color = 'blue')
plt.legend()
plt.xlabel('Score Category')
plt.ylabel('Total Questions')

plt.title('Number of questions with low and high score')

plt.show()

```
