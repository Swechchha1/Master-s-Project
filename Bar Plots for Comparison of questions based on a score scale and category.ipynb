```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as  sns
%matplotlib inline

df = pd.read_excel('surveys.xlsx')
df_quest = pd.read_excel('surveyquest.xlsx')
new_df = pd.merge(df, df_quest, on='question', how='inner')

# Creating different bar charts based on group of questions

new_df_gone = new_df[(new_df['group'] == 'Q1-Q6')] # 7 point scale (1-7)
new_df_gtwo = new_df[(new_df['group'] == 'Q7-Q14')] # 10 point scale (0-10)
new_df_gthree = new_df[(new_df['group'] == 'Q15-Q22')] # 4 point scale (1-4)


plt.figure(figsize=(10,7))
plt.title('Comparison of Questions 1 to 6 for low and high category on a 7 point scale', fontsize=20, color= 'b')
fig_one = sns.barplot(x='question',y='score', hue='category', palette=["C0", "C1"], data=new_df_gone)
fig_one.tick_params(axis = 'both', which = 'major', labelsize = 14)
plt.setp(fig_one.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(fig_one.get_legend().get_title(), fontsize='18') # for legend title
fig_one.set_ylabel('Scores', fontsize=15)
fig_one.set_xlabel('Question Number', fontsize=15)


plt.figure(figsize=(10,7))
plt.title('Comparison of Questions 7 to 14 for low and high category on a 10 point scale', fontsize=20, color= 'b')
fig_two = sns.barplot(x='question',y='score',hue='category', palette=["C1", "C0", ], data=new_df_gtwo)
fig_two.tick_params(axis = 'both', which = 'major', labelsize = 14)
plt.setp(fig_two.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(fig_two.get_legend().get_title(), fontsize='18') # for legend title
fig_two.set_ylabel('Scores', fontsize=15)
fig_two.set_xlabel('Question Number', fontsize=15)


plt.figure(figsize=(10,7))
plt.title('Comparison of Questions 15 to 22 for high category on a 4 point scale', fontsize=20, color= 'b')
fig_three = sns.barplot(x='question',y='score',hue='category',  palette=["C0", "C1"], data=new_df_gthree)
fig_three.tick_params(axis = 'both', which = 'major', labelsize = 14)
plt.setp(fig_three.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(fig_three.get_legend().get_title(), fontsize='18') # for legend title
fig_three.set_ylabel('Scores', fontsize=15)
fig_three.set_xlabel('Question Number', fontsize=15)

```
