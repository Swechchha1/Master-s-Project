```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output
import matplotlib.pyplot as plt
%matplotlib inline

"""
Analysis of each student's data over weeks for each question.
"""


df = pd.read_excel('surveys.xlsx') # No need to join this file with another file beacause catogery and groups are not required here

def unique_sorted_values(array):
    unique = array.unique().tolist()
    unique.sort()
    return unique

dropdown_quest = widgets.Dropdown(options = unique_sorted_values(df.question))
output_quest = widgets.Output()

def dropdown_quest_eventhandler(change):
    output_quest.clear_output()

    with output_quest:
            new_df=df[df.question == change.new]
            g = sns.relplot(x='week', y='score', data=new_df,kind='line', hue='surveycode', col = 'surveycode')
            display(g)
            plt.suptitle('Trend of a question no. ' +str(change.new) + ' over weeks for all students', fontsize=20, color= 'b', x = 0, y=1.2)
           
dropdown_quest.observe(dropdown_quest_eventhandler, names='value')        
display(dropdown_quest)
display(output_quest)


```
