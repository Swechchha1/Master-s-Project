```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
%matplotlib inline

df = pd.read_excel('surveys.xlsx')

is_question = df['question']==1
new_df = df[is_question]

g = sns.relplot(x='week', y='score', data=new_df,
            kind='line', hue='surveycode')


new_title = 'Trend of student with code "UDTK" for question 1'
g._legend.set_title(new_title)
```
